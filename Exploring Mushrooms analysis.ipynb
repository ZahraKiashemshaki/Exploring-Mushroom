```python
for index in range(0,len(df['Bruises'])):
    df['Bruises'][index] = str(df['Bruises'][index])
columns = df.columns.tolist()

for column in columns:
    print(column)
```

```python
for column in columns:
    sns.countplot(x=column, data=df)
```

```python
for column in columns:
    sns.countplot(x = df[column], data=df)
    plt.xticks(rotation=30, fontsize=10)
    plt.xlabel(column, fontsize=12)
    plt.show()
```

```python
for column in columns:
    sns.countplot(x = df[column], data=df, order=df[column].value_counts().index)
    plt.xticks(rotation=30, fontsize=10)
    plt.xlabel(column, fontsize=12)
    plt.title(column + " Value Counts")
    plt.show()
```

```python
prob =joint_prob = df.groupby(['Population', 'Habitat']).size() / len(df)
joint_prob = joint_prob.reset_index(name='Probability')
```

```python
labels = joint_prob.apply(lambda row: f"{row['Population']}-{row['Habitat']}", axis=1)
sizes = joint_prob['Probability']

plt.figure(figsize=(13, 13))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=200 )
plt.axis('equal')
plt.title('Joint Probability of Categorical Variables Population and Habitat')
plt.legend(loc="upper center")
plt.show()
```

```python
import scipy.stats as stats

contingency_table = pd.crosstab(df['Population'], df['Habitat'])
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)
print(f'Chi-Square Test:\nChi2: {chi2}\np-value: {p}\nDegrees of Freedom: {dof}')
```
